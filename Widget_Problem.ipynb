```python
from pulp import *
```

```python
model = LpProblem('Widget_Problem', LpMaximize)
```

```python
x = LpVariable('No.of_Widgetx: ', lowBound=0, cat='Integer')
y = LpVariable('No.of_Widgety: ', lowBound=0, cat='Integer')
```

```python
model += 4*x + 3*y
```

```python
model += x + y <= 48
model += y >= 2*x
model += x >= 0
model += y >= 0
```

```python
model.solve()
print('Status: ', LpStatus[model.status])
```

```python
for i in model.variables():
    print(i, '=', value(i))
print('Maximum_Profit: ', value(model.objective))
```
