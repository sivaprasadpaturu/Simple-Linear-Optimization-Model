```python
from pulp import *
```

```python
prob = LpProblem('Hammer_Problem', LpMinimize)
```

```python
A = LpVariable('Number of individual hammers of type A', lowBound=0, cat='Integer')
B = LpVariable('Number of individual hammers of type B', lowBound=0, cat='Integer')
```

```python
prob += 0.32 * A + 0.75 * B
```

```python
prob += A + 6*B >=70
```

```python
prob.solve()
print('Status:',LpStatus[prob.status])
```

```python
print("Optimal production from Factory A:", A.varValue)
print("Optimal production from Factory B:", B.varValue)
print("Minimum total cost:", value(prob.objective))
```
