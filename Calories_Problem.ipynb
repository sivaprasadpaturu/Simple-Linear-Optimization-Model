```python
from pulp import *
```

```python
# Problem
lp = LpProblem('calories_Problem', LpMaximize)
```

```python
# Variables
x1 = LpVariable(name='No_Of_Pushup', lowBound=0, cat ='Integer')
x2 = LpVariable(name='No_Of_Miles', lowBound=0, cat ='Integer')
```

```python
# objective Function
lp += 3 * x1 + 130 * x2
print(lp.objective)
```

```python
# Constraints
lp += 0.2 * x1 + 10 * x2 <= 10
print(lp.constraints)
```

```python
# solve LP
status = lp.solve()
print('Status:', LpStatus[lp.status])
```

```python
# Solution
for var in lp.variables():
    print(var, '=', value(var))
print('Calories_Burned = ', value(lp.objective))
```
