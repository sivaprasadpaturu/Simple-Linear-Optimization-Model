```python
from pulp import *
```

```python
model = LpProblem('Fitness_Problem', LpMinimize)
```

```python
x = LpVariable('No.of_Super_Bars_Consume: ', lowBound=0,cat='Integer')
y = LpVariable('No.of_Fantastic_Bars_Consume: ', lowBound=0,cat='Integer')
```

```python
model += 7*x + 4*y
```

```python
model += 20*x + 10*y >=140
model += 4*x + 6*y >=60
model += x >=0
model += y >=0
```

```python
model.solve()
print('Status: ', LpStatus[model.status])
```

```python
for i in model.variables():
    print(i, '=', value(i))
print('Minimum_Bars_Consume: ', value(model.objective))
```
