```python
from pulp import *
```

```python
# Problem
lp = LpProblem('Bakery_Problem', LpMaximize)
```

```python
# Variables
x1 = LpVariable(name='Chocolate_Cake', lowBound=0, cat='Integer')
x2 = LpVariable(name='Cheese_Cake', lowBound=0, cat ='Integer')
```

```python
# Objective Function
lp += 10 * x1 + 5 * x2
print(lp.objective)
```

```python
# Constraints
lp += 5 * x1 + x2 <= 90, 'Oven_Constraint'
lp += x1 + 10 * x2 <= 300, 'Food_Processor_Constraint'
lp += 4 * x1 + 6 * x2 <= 125, 'Bolier_Constraint'
print(lp.constraints)
```

```python
# Solve LP
status = lp.solve()
print('Status:', LpStatus[lp.status])
```

```python
# Solution
for var in lp.variables():
    print(var, '=', value(var))
print('OPT =', value(lp.objective))
```
