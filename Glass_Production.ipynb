```python
from pulp import *
```

```python
model = LpProblem('Maximize_Glass_Co._Profits', LpMaximize)
```

```python
x = LpVariable(name='No._of_Wine_glasses_produce: ',lowBound=0,cat='Integer')
y = LpVariable(name='No._of_Beer_glasses_produce: ',lowBound=0,cat='Integer')
```

```python
model+=5*x+4.5*y
```

```python
model+=6*x+5*y<=60
model+=10*x+20*y<=150
model+=x+y<=6
```

```python
model.solve()
print('Status:',LpStatus[model.status])
```

```python
for var in model.variables():
    print(var, '=', value(var))
print('Maximum_Profit: ', value(model.objective))
```
